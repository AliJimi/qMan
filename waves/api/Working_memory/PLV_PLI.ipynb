```python
import numpy as np
data_raw = np.load('D:\\jupyter_folder\\nurun_python_codes\\pre_Sa.npy') 
   
def analytic_signal(x):
    from scipy.fftpack import fft,ifft
    N = len(x)
    X = fft(x,N)
    h = np.zeros(N)
    h[0] = 1
    h[1:N//2] = 2*np.ones(N//2-1)
    h[N//2] = 1
    Z = X*h
    z = ifft(Z,N)
    return z
for jj in range (0,20):
    for j in range (0,20):
        ch_m=data_raw[jj,:]
        ch_n=data_raw[j,:]
        zm = analytic_signal(ch_m)
        zn = analytic_signal(ch_n)   
        plv=pli=np.zeros((21,21))
        gm=np.zeros(int(len(ch_m)))
        gn=gm
        radn=gn 
        radm=radn
        fe_n=radm
        fe_m=fe_n
        fe_mn=fe_m
        for i in range(0,len(ch_m)-1):
            gm[i]=zm[i]/ch_m[i]
            gn[i]=zn[i]/ch_n[i]
            radm[i]=np.angle(gm[i],deg=False)
            radn[i]=np.angle(gn[i],deg=False)
            fe_m[i]=np.arctan(radm[i])
            fe_n[i]=np.arctan(radn[i])
            fe_mn[i]=fe_m[i]-fe_n[i]
            plv[jj,j]=np.sqrt((((np.mean(np.cos(fe_mn[i])))**2)+(np.mean(np.sin(fe_mn[i])))**2))
            pli[jj,j]=np.mean(np.sign(fe_mn))
print(plv,pli)
```

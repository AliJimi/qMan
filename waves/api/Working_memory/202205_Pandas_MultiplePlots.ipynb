```python
import numpy as np
import seaborn as sns
import pandas as pd
import random
```

```python
# random sample generation
features = np.random.random_sample(size = (50,100))
subj_cog_test = np.random.random_sample(50)
```

```python
# Name the feature 
feature_names = []
for i in range(len(features[0])):
    feature_names.append('feature' + str(i+1))
```

```python
# create data fram with features as column and rows as subjects
df = pd.DataFrame(features, columns= feature_names)
```

```python
# add cognitive test as a column
df['subjectCogTestScore'] = subj_cog_test
```

```python
df = df.melt('subjectCogTestScore', var_name='cols', value_name='featureVals')
```

```python
g = sns.FacetGrid(df, col= "cols",   height=2.5, col_wrap=10)
g.map(sns.scatterplot, "featureVals","subjectCogTestScore")
for ax in g.axes.flat:
    ax.axline((0, 0), slope=1, c="0.5", ls="--", zorder=0)
```

```python
import scipy.io
import scipy.signal as signal
from scipy.signal import butter, lfilter
import numpy as np
import pandas as pd
import mne
```

```python
#load raw data
data_raw = scipy.io.loadmat('y.mat')
f=data_raw['y']
g=np.array((f))

video_length=80 #seconds
baseline_length=10 #seconds
sampling_rate=500
trial_length=(video_length+baseline_length)*sampling_rate

number_trials=10 #number of videos
final_data_length= trial_length*number_trials

#extract the events row
group=g[50]
##extract the part of data that participants are watching video
smpls=group.size
indx=np.empty(smpls)
for n in range(1,smpls):
    if ((group[n-1]==0 and group[n]==1) or (group[n-1]==3 and group[n]==1)):
     indx[n-1]=n
    else:
     indx[n-1]=0
    
t=np.argwhere(indx>0)
time=indx[t]
num_trial=time.size
uu=np.empty((52, final_data_length))   

for num in range(0,number_trials):
    ok=int(time[num])
    uu[:,((num)*trial_length):((num+1)*trial_length)]= g[:,ok:(ok+trial_length)]
    
#extracting EEG, fNIRS optic and fNIRS oxygen and heart rate
fnirs_ox=uu[1:17]
fnirs_opitc=uu[17:33]
eeg_data=uu[33:49,:]
heart_rate=uu[51,:]
```

```python
import mayavi
eee=eeg_data[:,0:10000]
gg=eee.tostring()
coreg=mne.gui.coregistration(gg)
np.save('gggg', gg)
```
